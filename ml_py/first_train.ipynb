```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0
def forward(x):
    return x*w

def loss(x,y):
    y_predict = forward(x)
    return (y_predict - y)*(y_predict - y)

def gradient(x,y):
    return 2*x*(x*w-y)

print("my prediction before: ", 4, forward(4))

for epoch in range(20):
    for x_val,y_val in zip(x_data,y_data):
        grad = gradient(x_val, y_val)
        w = w - grad * 0.01
        print("\tgrad: " , x_val, y_val, round(grad, 2))
        l=loss(x_val,y_val)
    print("progress: " , epoch,"w=",round(w,2),"loss=", round(l,2))
print("my new prediction", forward(4))
```
